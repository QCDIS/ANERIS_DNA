```python
# PEMASequenceRetriever
# =====

# test in terminal
# python /home/jovyan/library/enaBrowserTools/python3_1.6/enaDataGet.py -f fastq -d /tmp/data/ena ERR12541385

# -----
dir_cwd = os.getcwd()

# dir_tmp = "/tmp/data"
dir_tmp = os.path.join(dir_cwd, "ToDel")

dir_lib = os.path.join(dir_cwd, "library")

# -----
dir_lib_ena = os.path.join(dir_lib, "enaBrowserTools", "python3_1.6")

dir_tmp_ena = os.path.join(dir_tmp, "1_ena")
dir_tmp_seq = os.path.join(dir_tmp, "2_seq")

try:
    os.makedirs(dir_tmp_ena)
except FileExistsError:
    # directory already exists
    pass

try:
    os.makedirs(dir_tmp_seq)
except FileExistsError:
    # directory already exists
    pass

# -----
file_lib_ena = os.path.join(dir_lib_ena, conf_fname_lib_ena)
file_seq_zip = os.path.join(dir_tmp_seq, param_fname_seq_zip)

# -----
genes_list = param_gene_sequences.split(',') if len(param_gene_sequences) else []

if (len(genes_list) == 0):
    raise RuntimeError("At least 1 gene sequence must be provided")

for gene in genes_list:
    # Run ENA browser tools
    # os.system(f"python {file_lib_ena} -f fastq -d {dir_tmp_ena} {gene}")
    os.system(f"python {file_lib_ena} -f fastq -d {dir_tmp_ena} -c {gene}")
    
    if len(os.listdir(os.path.join(dir_tmp_ena, gene))) != 2:
        raise RuntimeError(f"Expected 2 files for gene `{gene}`. Found {len(os.listdir(dir_tmp_ena))} @ {dir_tmp_ena}")

file_fastq_list = glob.glob(os.path.join(dir_tmp_ena, "*", "*.fastq.gz"))

with zipfile.ZipFile(file_seq_zip, "w") as zipObj:
    for file_fastq in file_fastq_list:
        zipObj.write(file_fastq, os.path.basename(file_fastq))
        print(f"Compress gene file {file_fastq}")
print(f"Generated compressed sequencing files {file_fastq}")

if not os.path.isfile(file_seq_zip):
    raise FileNotFoundError(str(file_seq_zip) + " is missing")
```
